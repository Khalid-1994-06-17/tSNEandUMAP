```python
from sklearn.datasets import make_s_curve
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate S-curve dataset
X, color = make_s_curve(n_samples=1000, noise=0.1, random_state=42)

# Plot the original 3D S-curve
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap='viridis', s=10)
ax.set_title("3D S-Curve Dataset")
plt.show()
```

```python
from sklearn.decomposition import PCA

# Apply PCA to reduce the S-Curve to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot PCA output
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=color, cmap='viridis', s=10)
plt.title("PCA on S-Curve Dataset")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

```python
from sklearn.datasets import make_swiss_roll, make_moons, make_blobs, load_iris
from sklearn.datasets import load_digits, load_wine, load_breast_cancer
from sklearn.decomposition import PCA, KernelPCA
from sklearn.manifold import TSNE
import umap
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score, adjusted_rand_score, normalized_mutual_info_score
import pandas as pd

# Function to apply dimensionality reduction methods
def apply_reduction(method, X, **kwargs):
    if method == "PCA":
        reducer = PCA(n_components=2)
    elif method == "Kernel PCA":
        reducer = KernelPCA(n_components=2, kernel=kwargs.get("kernel", "rbf"), gamma=kwargs.get("gamma", 0.1))
    elif method == "t-SNE":
        reducer = TSNE(n_components=2, random_state=42)
    elif method == "UMAP":
        reducer = umap.UMAP(n_components=2, random_state=42)
    else:
        raise ValueError("Invalid reduction method")
    return reducer.fit_transform(X)

# Function to evaluate clustering performance
def evaluate_clustering(X_reduced, labels_true):
    gmm = GaussianMixture(n_components=len(set(labels_true)), random_state=42)
    gmm_labels = gmm.fit_predict(X_reduced)
    return {
        "Silhouette": silhouette_score(X_reduced, gmm_labels),
        "ARI": adjusted_rand_score(labels_true, gmm_labels),
        "NMI": normalized_mutual_info_score(labels_true, gmm_labels)
    }

# Prepare datasets
datasets = {
    "Moons": make_moons(n_samples=500, noise=0.05, random_state=42),
    "Iris": load_iris(return_X_y=True),
    "Wine": load_wine(return_X_y=True),
    "Digits": load_digits(return_X_y=True),
    "Breast Cancer": load_breast_cancer(return_X_y=True)
}

methods = ["PCA", "Kernel PCA", "t-SNE", "UMAP"]
results = []

# Iterate over datasets
for dataset_name, (X, y) in datasets.items():
    for method in methods:
        # Apply dimensionality reduction
        if method == "Kernel PCA":
            X_reduced = apply_reduction(method, X, kernel="rbf", gamma=0.1)
        else:
            X_reduced = apply_reduction(method, X)
        
        # Evaluate clustering
        scores = evaluate_clustering(X_reduced, y)
        results.append({
            "Dataset": dataset_name,
            "Method": method,
            **scores
        })

# Convert results to a DataFrame
df_results = pd.DataFrame(results)
print(df_results)
```

```python
import matplotlib.pyplot as plt

# Creating a grouped bar chart for visualization with numbers and grid
fig, ax = plt.subplots(figsize=(16, 8))

# Define bar width and positions
bar_width = 0.2
datasets = df_results["Dataset"].unique()
methods = df_results["Method"].unique()
x_positions = range(len(datasets))

# Plot each method's scores
for i, method in enumerate(methods):
    scores = df_results[df_results["Method"] == method]["ARI"]
    bars = ax.bar(
        [x + i * bar_width for x in x_positions],
        scores,
        bar_width,
        label=method,
    )
    # Add text annotations for the scores
    for bar, score in zip(bars, scores):
        ax.text(
            bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 0.01,
            f"{score:.2f}",
            ha="center",
            va="bottom",
            fontsize=10
        )

# Add labels, title, grid, and legend
ax.set_title("ARI Scores by Dataset and Method", fontsize=16)
ax.set_xlabel("Dataset", fontsize=12)
ax.set_ylabel("ARI Score", fontsize=12)
ax.set_xticks([x + bar_width for x in x_positions])
ax.set_xticklabels(datasets)
ax.legend(title="Method")
ax.grid(axis="y", linestyle="--", alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons, load_iris, load_wine, load_digits, load_breast_cancer
from sklearn.decomposition import PCA, KernelPCA
from sklearn.manifold import TSNE
import umap
from sklearn.mixture import GaussianMixture
import numpy as np

# Prepare datasets
datasets = {
    "Moons": make_moons(n_samples=500, noise=0.05, random_state=42),
    "Iris": load_iris(return_X_y=True),
    "Wine": load_wine(return_X_y=True),
    "Digits": load_digits(return_X_y=True),
    "Breast Cancer": load_breast_cancer(return_X_y=True)
}

methods = ["PCA", "Kernel PCA", "t-SNE", "UMAP"]

# Function to apply dimensionality reduction methods
def apply_reduction(method, X, **kwargs):
    if method == "PCA":
        reducer = PCA(n_components=2)
    elif method == "Kernel PCA":
        reducer = KernelPCA(n_components=2, kernel=kwargs.get("kernel", "rbf"), gamma=kwargs.get("gamma", 0.1))
    elif method == "t-SNE":
        reducer = TSNE(n_components=2, random_state=42)
    elif method == "UMAP":
        reducer = umap.UMAP(n_components=2, random_state=42)
    else:
        raise ValueError("Invalid reduction method")
    return reducer.fit_transform(X)

# Function to plot clusters
def plot_clusters(X_reduced, labels, dataset_name, method, ax):
    sns.scatterplot(x=X_reduced[:, 0], y=X_reduced[:, 1], hue=labels, palette="tab10", ax=ax, s=20)
    ax.set_title(f"{dataset_name} - {method}")
    ax.set_xlabel("Component 1")
    ax.set_ylabel("Component 2")
    ax.legend([], [], frameon=False)  # Remove legend for clarity

# Plot the clusters for each dataset and method
fig, axes = plt.subplots(len(datasets), len(methods), figsize=(20, len(datasets) * 4))

for i, (dataset_name, (X, y)) in enumerate(datasets.items()):
    for j, method in enumerate(methods):
        # Apply dimensionality reduction
        if method == "Kernel PCA":
            X_reduced = apply_reduction(method, X, kernel="rbf", gamma=0.1)
        else:
            X_reduced = apply_reduction(method, X)
        
        # Gaussian Mixture clustering
        gmm = GaussianMixture(n_components=len(np.unique(y)), random_state=42)
        gmm_labels = gmm.fit_predict(X_reduced)
        
        # Plot clusters
        plot_clusters(X_reduced, gmm_labels, dataset_name, method, axes[i, j])

plt.tight_layout()
plt.show()
```
